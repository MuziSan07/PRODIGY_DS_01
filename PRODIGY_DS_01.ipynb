```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Iris.csv")
df
```

```python
sns.countplot(x='Species', data=df, palette='viridis')
plt.title('Distribution of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(df['SepalLengthCm'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Sepal Length in Iris Dataset')
plt.xlabel('SepalLengthCm')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['SepalWidthCm'], bins=20, kde=True, color='blue', edgecolor='black')
plt.title('Distribution of Sepal Width in Iris Dataset')
plt.xlabel('SepalWidthCm')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['PetalLengthCm'], bins=20, kde=True, color='purple', edgecolor='black')
plt.title('Distribution of Petal Length	in Iris Dataset')
plt.xlabel('PetalLengthCm	')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['PetalWidthCm'], bins=20, kde=True, color='Pink', edgecolor='black')
plt.title('Distribution of Petal Width in Iris Dataset')
plt.xlabel('PetalWidthCm')
plt.ylabel('Frequency')
plt.show()
```
